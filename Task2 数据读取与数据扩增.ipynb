## 2. 数据扩增方法
**数据扩增的概念**

数据扩增是指不实际增加原始数据，只是对原始数据做一些变换，从而创造出更多的数据。

**数据扩增的目的**

数据扩增的目的是增加数据量、丰富数据多样性、提高模型的泛化能力。

**数据扩增的基本原则**

不能引入无关的数据
扩增总是基于先验知识的，对于不同的任务和场景，数据扩增的策略也会不同。
扩增后的标签保持不变
