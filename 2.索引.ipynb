### 1. loc方法，iloc方法，[]操作符
loc表示标签索引  
iloc表示位置索引  
[]可以很方便的进行一些操作  
  
注意点：  
loc中使用的切片全部包含**右端点**   
iloc中的切片不包含右端点  
  
[]只输一个值或用：连接两个数，取的是某行或者某几行    

**总结**:  
1. **df.loc[a:b:s, c]**  
逗号左边代表取行的范围，右边代表列    
a 为起始的索引值  
b 为截止的索引值  
s 为步长，即从a到b的范围，以s步长取数，s可以为负值  
c 为列名，可以单个或多个列  
2. 本质上说，loc中能传入的只有布尔列表和索引子集构成的列表
3. **df.iloc[a:b:s, c]**  
和loc的一样，只不过iloc中的ab代表的是位置
3. iloc中接收的参数只能为整数或整数列表或布尔列表，不能使用布尔Series，如果要用就必须把values拿出来

### 2.布尔索引
**布尔符号**   
& 代表 和 and  
| 代表 或 or  
~ 取反  
loc和iloc和[]中相应位置都能使用布尔列表选择：  
  
**isin方法**  
可以指定某列具体取哪些值

```python
# 布尔符号
#df[df['Gender']=='F' & df['Address']=='street_2'].head() #报错，条件必须用括号（）包起来
df[(df['Gender']=='F') & (df['Address']=='street_2')].head()

df[(df['Math'] > 85) | (df['Address']=='street_7')].head()

df[~((df['Math'] > 85) & (df['Address']=='street_7'))].head() # and取反 等效于 (df['Math'] <= 85) | (df['Address'] != 'street_7')
df[~((df['Math'] > 85) | (df['Address']=='street_7'))].head() # or取反 等效于 (df['Math'] <= 85) and (df['Address'] != 'street_7')
```

```python
# loc和[]中相应位置都能使用布尔列表选择：
df.loc[df['Math']>60, (df[:8]['Address']=='street_6').values].head()
#如果不加values就会索引对齐发生错误，Pandas中的索引对齐是一个重要特征，很多时候非常实用
#但是若不加以留意，就会埋下隐患

#df.loc[df['Math']>60, (df[:8]['Address']=='street_6')].head() #报错
```

```python
# isin方法
# isin() 括号内需要是list
df[df['Address'].isin(['street_1','street_4']) & df['Physics'].isin(['A','A+'])]

# 还可以用字典的方式写
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(1)]
#all与&的思路是类似的，其中的1代表按照跨列方向判断是否全为True

# 不写all(1)
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']})].head()
# df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}) # 一串 true or false list
```

### 4.区间索引
此处介绍并不是说只能在单级索引中使用区间索引，只是作为一种特殊类型的索引方式，在此处先行介绍
  
#### a）利用interval_range方法

```python

pd.interval_range(start=0,end=5)
#closed参数可选'left''right''both''neither'，默认左开右闭, 从0开始

pd.interval_range(start=0, periods=8, freq=5)
#periods参数控制区间个数，freq控制步长
```

#### b） 利用cut将数值列转为区间,作为为元素的分类变量
例如统计数学成绩的区间情况

```python
math_interval = pd.cut(df['Math'], bins=[0,40,60,80,100]) # 默认左开右闭
#注意，如果没有类型转换，此时并不是区间类型，而是category类型
math_interval.head()
```

#### c） 区间索引的选取


```python
df_i = df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval')
df_i.head()

# rsuffix 指定后缀，不然列名重复会报错
df.join(math_interval, rsuffix='_interval')  #表示df 和math_interval拼接
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']] # 拼接后取math 和 math_interval两列
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index() # 重新设置索引，从0开始
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval') 
#将索引设置为math_interval， 设为索引后， 该列消失
```

```python
# 包含单个值
df_i.loc[65] # 判定区间索引是否包含65
df_i.loc[[65,90]] # 判定区间索引是否包含65或90，包含就取出

# 如果想要判断是否包含某个区间，先要把分类变量转为区间变量，再使用overlap方法：
#df_i.loc[pd.Interval(70,75)].head() 报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70, 85))].head()
```

```python
# 可以选择缺失值的填充方法：fill_value和method（bfill/ffill/nearest），其中method参数必须索引单调
#bfill表示用所在索引1206的后一个有效行填充，ffill为前一个有效行，nearest是指最近的

# 1206原本不存在，现在选择用前 后 最近的标准填充
df.reindex(index=[1101,1203,1206,2402],method='bfill')
df.reindex(index=[1101,1203,1206,2402],method='ffill')
df.reindex(index=[1101,1203,1206,2402],method='nearest')
#数值上1205比1301更接近1206，因此用前者填充
```

```python
# reindex_like的作用为生成一个横纵索引完全与参数列表一致的DataFrame，数据使用被调用的表
df_temp = pd.DataFrame({'Weight':np.zeros(5),
                        'Height':np.zeros(5),
                        'ID':[1101,1104,1103,1106,1102]}).set_index('ID')
df_temp
df[0:5][['Weight','Height']] # 取df前5个
df_temp.reindex_like(df[0:5][['Weight','Height']]) 
# 将df_temp的索引设置为df的 1101-1105， 对应的值是索引在df_temp中对应的值，所以如果某个索引在df_temp中不存在，就显示为NaN
```

### 3.query 函数
df.query（expr，inplace = False，** kwargs ）  
使用布尔表达式查询帧的列
 
**参数**：  
expr：str要评估的查询字符串。你可以在环境中引用变量，在它们前面添加一个'@'字符 。@a + b  
inplace=False：是否修改数据或返回副本  
kwargs：dict关键字参数  
 
**返回**：DataFrame  
 
**注意**：  
默认修改Python语法 '&' / 'and' 和 '|' / 'or' 位运算符优先级高于布尔表达式，不同于Python。  
关键字参数parser='python'执行Python评估。  
engine='python' 用Python本身作为后端来传递评估表达式。不建议效率低。  
 
默认实例df.index和 df.columns属性 DataFrame放在查询命名空间中，  
这允许您将框架的索引和列视为框架中的列。标识符index用于帧索引;   
您还可以使用索引的名称在查询中标识它。  

```python
# query函数中的布尔表达式中，下面的符号都是合法的：行列索引名、字符串、and/not/or/&/|/~/not in/in/==/!=、四则运算符
df.query('(Address in ["street_6","street_7"]) & (Weight>(70+10)) & (ID in [1303,2304,2402])')
```

## 六、抽样函数
这里的抽样函数指的就是sample函数  
  
**df.sample(n, frac, replace, axis, weights)**    
n为样本量  
frac为抽样比  
replace为是否放回  
axis为抽样维度，默认为0，即抽行，1为列  
weights为样本权重，自动归一化  

```python
df.sample(n=5) # 随机抽5行
df.sample(frac=0.05) # 随机抽5%
df.sample(n=df.shape[1],replace=True) # true为有放回的抽样，即结果有可能是由重复数据的
df.shape[0] # 行数，35
df.shape[1] # 列数，8
df.sample(n=3,axis=1).head()
```

```python
# weight是设定的权重即概率大小，并不代表一定就抽到概率大的
# weight参数提供语法糖，只要你给1 2 3 4 5的绝对量，自动会把你输入转成归一化之后的概率进行抽样

np.random.rand(df.shape[0]) # 设定权重
df.sample(n=3,weights=np.random.rand(df.shape[0]))

#以某一列为权重，这在抽样理论中很常见
#抽到的概率与Math数值成正比
df.sample(n=3,weights=df['Math']).head()
```
