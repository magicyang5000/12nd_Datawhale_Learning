### 1. loc方法，iloc方法，[]操作符
loc表示标签索引  
iloc表示位置索引  
[]可以很方便的进行一些操作  
  
注意点：  
loc中使用的切片全部包含**右端点**   
iloc中的切片不包含右端点  
  
[]只输一个值或用：连接两个数，取的是某行或者某几行    

**总结**:  
1. **df.loc[a:b:s, c]**  
逗号左边代表取行的范围，右边代表列    
a 为起始的索引值  
b 为截止的索引值  
s 为步长，即从a到b的范围，以s步长取数，s可以为负值  
c 为列名，可以单个或多个列  
2. 本质上说，loc中能传入的只有布尔列表和索引子集构成的列表
3. **df.iloc[a:b:s, c]**  
和loc的一样，只不过iloc中的ab代表的是位置
3. iloc中接收的参数只能为整数或整数列表或布尔列表，不能使用布尔Series，如果要用就必须把values拿出来

### 2.布尔索引
**布尔符号**   
& 代表 和 and  
| 代表 或 or  
~ 取反  
loc和iloc和[]中相应位置都能使用布尔列表选择：  
  
**isin方法**  
可以指定某列具体取哪些值

```python
# 布尔符号
#df[df['Gender']=='F' & df['Address']=='street_2'].head() #报错，条件必须用括号（）包起来
df[(df['Gender']=='F') & (df['Address']=='street_2')].head()

df[(df['Math'] > 85) | (df['Address']=='street_7')].head()

df[~((df['Math'] > 85) & (df['Address']=='street_7'))].head() # and取反 等效于 (df['Math'] <= 85) | (df['Address'] != 'street_7')
df[~((df['Math'] > 85) | (df['Address']=='street_7'))].head() # or取反 等效于 (df['Math'] <= 85) and (df['Address'] != 'street_7')
```

```python
# loc和[]中相应位置都能使用布尔列表选择：
df.loc[df['Math']>60, (df[:8]['Address']=='street_6').values].head()
#如果不加values就会索引对齐发生错误，Pandas中的索引对齐是一个重要特征，很多时候非常实用
#但是若不加以留意，就会埋下隐患

#df.loc[df['Math']>60, (df[:8]['Address']=='street_6')].head() #报错
```

```python
# isin方法
# isin() 括号内需要是list
df[df['Address'].isin(['street_1','street_4']) & df['Physics'].isin(['A','A+'])]

# 还可以用字典的方式写
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(1)]
#all与&的思路是类似的，其中的1代表按照跨列方向判断是否全为True

# 不写all(1)
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']})].head()
# df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}) # 一串 true or false list
```

### 4.区间索引
此处介绍并不是说只能在单级索引中使用区间索引，只是作为一种特殊类型的索引方式，在此处先行介绍
  
#### a）利用interval_range方法

```python

pd.interval_range(start=0,end=5)
#closed参数可选'left''right''both''neither'，默认左开右闭, 从0开始

pd.interval_range(start=0, periods=8, freq=5)
#periods参数控制区间个数，freq控制步长
```

#### b） 利用cut将数值列转为区间,作为为元素的分类变量
例如统计数学成绩的区间情况

```python
math_interval = pd.cut(df['Math'], bins=[0,40,60,80,100]) # 默认左开右闭
#注意，如果没有类型转换，此时并不是区间类型，而是category类型
math_interval.head()
```

#### c） 区间索引的选取


```python
df_i = df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval')
df_i.head()

# rsuffix 指定后缀，不然列名重复会报错
df.join(math_interval, rsuffix='_interval')  #表示df 和math_interval拼接
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']] # 拼接后取math 和 math_interval两列
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index() # 重新设置索引，从0开始
df.join(math_interval, rsuffix='_interval')[['Math','Math_interval']].reset_index().set_index('Math_interval') 
#将索引设置为math_interval， 设为索引后， 该列消失
```

```python
# 包含单个值
df_i.loc[65] # 判定区间索引是否包含65
df_i.loc[[65,90]] # 判定区间索引是否包含65或90，包含就取出

# 如果想要判断是否包含某个区间，先要把分类变量转为区间变量，再使用overlap方法：
#df_i.loc[pd.Interval(70,75)].head() 报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70, 85))].head()
```

### 1. 多级索引的创建


```python
# 通过from_tuple或from_arrays创建
## 直接创建
tuples = [('A','a'),('A','b'),('B','a'),('B','b')]
mul_index = pd.MultiIndex.from_tuples(tuples, names=('Upper', 'Lower'))
mul_index

pd.DataFrame({'Score':['perfect','good','fair','bad']},index=mul_index)

## 利用zip创建元组
L1 = list('AABB')
L2 = list('abab')
zip(L1,L2) # 一一对应
tuples = list(zip(L1,L2))
tuples
mul_index = pd.MultiIndex.from_tuples(tuples, names=('Upper', 'Lower'))
pd.DataFrame({'Score':['perfect','good','fair','bad']},index=mul_index)

##通过array创建
arrays = [['A','a'],['A','b'],['B','a'],['B','b']]
mul_index = pd.MultiIndex.from_tuples(arrays, names=('Upper', 'Lower'))
pd.DataFrame({'Score':['perfect','good','fair','bad']},index=mul_index)
```

```python
# 通过from_product
## 两两相乘
L1 = ['A','B'] 
L2 = ['a', 'b']
pd.MultiIndex.from_product([L1,L2], names=('Upper', 'Lower'))
```

### 2. 多层索引切片


```python
df_using_mul.head()

# 一般切片
#df_using_mul.loc['C_2','street_5'] #当索引不排序时，单个索引会报出性能警告
df_using_mul.index.is_lexsorted()  #该函数检查是否排序
df_using_mul.sort_index().loc['C_2','street_5']
df_using_mul.sort_index().index.is_lexsorted()
df_using_mul.sort_index().loc[('C_2','street_6'):('C_3','street_4')]

df_using_mul.sort_index().loc[('C_2','street_7'):'C_3'].head() #非元组也是合法的，表示选中该层所有元素
```

### 3. 多层索引中的slice对象

```python
L1,L2 = ['A','B','C'],['a','b','c']
mul_index1 = pd.MultiIndex.from_product([L1,L2],names=('Upper', 'Lower'))
mul_index1
L3,L4 = ['D','E','F'],['d','e','f']
mul_index2 = pd.MultiIndex.from_product([L3,L4],names=('Big', 'Small'))
mul_index2
df_s = pd.DataFrame(np.random.rand(9,9),index=mul_index1,columns=mul_index2) # 不仅可以设置列index，还可以设置行index
df_s

# 有啥用？
```

```python
idx=pd.IndexSlice
df_s.loc[idx['B':,df_s['D']['d']>0.3],idx[df_s.sum()>4]]
# 列索引， 从B开始 到最后
# 行索引， 限制D下面的d 必须大于0.3
# df_s.sum()默认为对列求和，因此返回一个长度为9的数值列表
idx[df_s.sum()>4] 
```

### 2. reindex和reindex_like

```python
# reindex是指重新索引，它的重要特性在于索引对齐，很多时候用于重新排序
df.head()
df.reindex(index=[1101,1203,1206,2402])
df.reindex(columns=['Height','Gender','Average']).head()
```

```python
# 可以选择缺失值的填充方法：fill_value和method（bfill/ffill/nearest），其中method参数必须索引单调
#bfill表示用所在索引1206的后一个有效行填充，ffill为前一个有效行，nearest是指最近的

# 1206原本不存在，现在选择用前 后 最近的标准填充
df.reindex(index=[1101,1203,1206,2402],method='bfill')
df.reindex(index=[1101,1203,1206,2402],method='ffill')
df.reindex(index=[1101,1203,1206,2402],method='nearest')
#数值上1205比1301更接近1206，因此用前者填充
```

```python
# reindex_like的作用为生成一个横纵索引完全与参数列表一致的DataFrame，数据使用被调用的表
df_temp = pd.DataFrame({'Weight':np.zeros(5),
                        'Height':np.zeros(5),
                        'ID':[1101,1104,1103,1106,1102]}).set_index('ID')
df_temp
df[0:5][['Weight','Height']] # 取df前5个
df_temp.reindex_like(df[0:5][['Weight','Height']]) 
# 将df_temp的索引设置为df的 1101-1105， 对应的值是索引在df_temp中对应的值，所以如果某个索引在df_temp中不存在，就显示为NaN
```

```python
# 如果df_temp单调还可以使用method参数：
df_temp = pd.DataFrame({'Weight':range(5),
                        'Height':range(5),
                        'ID':[1101,1104,1103,1106,1102]}).set_index('ID').sort_index()
df_temp
df_temp.reindex_like(df[0:5][['Weight','Height']],method='ffill')
#可以自行检验这里的1105的值是否是由bfill规则填充
df_temp.reindex_like(df[0:5][['Weight','Height']],method='bfill')
```

### 3. set_index和reset_index

```python
# set_index：将某些列作为索引
df.head()
df.set_index('Class').head()

# 利用append参数可以将当前索引维持不变
df.set_index('Class',append=True).head()

# 设置多级索引
df.set_index([pd.Series(range(df.shape[0])),pd.Series(np.ones(df.shape[0]))]).head()
```

```python
# reset_index方法，它的主要功能是将索引重置
# 默认状态直接恢复到自然数索引：
df.reset_index().head()

# 用level参数指定哪一层被reset，用col_level参数指定set到哪一层：
L1,L2 = ['A','B','C'],['a','b','c']
mul_index1 = pd.MultiIndex.from_product([L1,L2],names=('Upper', 'Lower'))
L3,L4 = ['D','E','F'],['d','e','f']
mul_index2 = pd.MultiIndex.from_product([L3,L4],names=('Big', 'Small'))
df_temp = pd.DataFrame(np.random.rand(9,9),index=mul_index1,columns=mul_index2)
df_temp.head()

df_temp1 = df_temp.reset_index(level=1,col_level=1)
df_temp1.head()

df_temp1.columns #看到的确插入了level2
df_temp1.index #最内层索引被移出
```

### 1.where 函数

这里的抽样函数指的就是sample函数  
  
**df.sample(n, frac, replace, axis, weights)**    
n为样本量  
frac为抽样比  
replace为是否放回  
axis为抽样维度，默认为0，即抽行，1为列

#### 【练习一】 现有一份关于UFO的数据集，请解决下列问题：

```python
UFO = pd.read_csv('./data/UFO.csv')
UFO.info()
UFO.head()
```

##### （a）在所有被观测时间超过60s的时间中，哪个形状最多？
##### （b）对经纬度进行划分：-180°至180°以30°为一个经度划分，-90°至90°以18°为一个维度划分，请问哪个区域中报告的UFO事件数量最多？

```python
# a 在所有被观测时间超过60s的时间中，哪个形状最多？
# 答案是 light的形状最多

# 先筛选出观测时间>60s的
# 在对筛选结果，对shape字段做类型统计，直接value_counts()
UFO[UFO['duration (seconds)'] >= 60.0]['shape'].value_counts()


# b.对经纬度进行划分：-180°至180°以30°为一个经度划分，-90°至90°以18°为一个维度划分，请问哪个区域中报告的UFO事件数量最多？
# 答案：((30, 60], (-90, -72]) 报告事件最多，达30952次

# 先对数据进行筛选，经度±180以外 纬度±90以外的去掉
data = UFO[(UFO['latitude'] >= -180.0) & (UFO['latitude'] <= 180.0) & (UFO['longitude'] >= -90.0) & (UFO['longitude'] <= 90.0)]
# 用cut函数构建区间
latitude_interval = pd.cut(data['latitude'], bins= [-180,-150,-120,-90,-60,-30,0,30,60,90,120,150,180])
longitude_interval = pd.cut(data['longitude'], bins= [-90,-72,-54,-36,-18,0,18,36,54,72,90])
# 将区间匹配到数据集上
data = data.join(latitude_interval, rsuffix='_interval')
data = data.join(longitude_interval, rsuffix='_interval')
# 将两个区间列变为索引，并对索引value_counts
# 思路二 其实可以单纯将两个字段拼成新的列，然后对新列 value_counts()
data.set_index(['latitude_interval','longitude_interval']).index.value_counts() 
```
