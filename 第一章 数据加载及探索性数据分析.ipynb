# 2. pandas基础

## Series VS Dataframe

区别：  
series，只是一个一维数据结构，它由index和value组成。  
dataframe，是一个二维结构，除了拥有index和value之外，还拥有column。  
联系：  
dataframe由多个series组成，无论是行还是列，单独拆分出来都是一个series。

# 3. 探索性数据分析

## 排序

```python
df = pd.DataFrame(np.arange(8).reshape((2, 4)), index=[2,1], columns=['d', 'a', 'b', 'c']) 
df
```

```python
# 指定列排序, ascending 升序降序
df.sort_values(by='a', axis=0,ascending=False)

# 指定行排序
df.sort_values(by=1, axis=1, ascending=False)


# 多列排序
df.sort_values(by=['a','b'] , ascending=False)
```

```python
df_train.sort_values(by=['票价', '年龄'], ascending=False).head(3)
```

## dataframe算术计算

```python
frame_a = pd.DataFrame(np.arange(9.).reshape(3, 3),
                     columns=['a', 'b', 'c'],
                     index=['one', 'two', 'three'])
frame_b = pd.DataFrame(np.arange(12.).reshape(4, 3),
                     columns=['a', 'e', 'c'],
                     index=['first', 'one', 'two', 'second'])
frame_a
frame_b
```

```python
frame_a + frame_b
# 对应的行和列的值会相加，没有对应的会变成空值NaN。
```
