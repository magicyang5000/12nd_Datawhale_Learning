## 【练习一】 现继续使用第四章中的地震数据集，请解决以下问题：
### （a）现在将深度分为七个等级：[0,5,10,15,20,30,50,np.inf]，请以深度等级Ⅰ,Ⅱ,Ⅲ,Ⅳ,Ⅴ,Ⅵ,Ⅶ为索引并按照由浅到深的顺序进行排序。
### （b）在（a）的基础上，将烈度分为4个等级：[0,3,4,5,np.inf]，依次对南部地区的深度和烈度等级建立多级索引排序。

```python
Earthquake = pd.read_csv('data/Earthquake.csv')
Earthquake.info()
```

```python
#（a）现在将深度分为七个等级：[0,5,10,15,20,30,50,np.inf]，请以深度等级Ⅰ,Ⅱ,Ⅲ,Ⅳ,Ⅴ,Ⅵ,Ⅶ为索引并按照由浅到深的顺序进行排序。

## 分等级
# pd.cut(Earthquake['深度'], [0,5,10,15,20,30,50,np.inf], right=False)

## 直接将等级赋值
Earthquake['Dep Levl'] = pd.cut(Earthquake['深度'], [0,5,10,15,20,30,50,np.inf], right=False, labels=['Ⅰ','Ⅱ','Ⅲ','Ⅳ','Ⅴ','Ⅵ','Ⅶ'])\
                        .astype('category').cat.as_ordered()
Earthquake.set_index('Dep Levl').sort_index()
```

```python
#（b）在（a）的基础上，将烈度分为4个等级：[0,3,4,5,np.inf]，依次对南部地区的深度和烈度等级建立多级索引排序。
Earthquake['CSIS Levl'] = pd.cut(Earthquake['烈度'], [0,3,4,5,np.inf], right=False, labels=['Ⅰ','Ⅱ','Ⅲ','Ⅳ'])\
                        .astype('category').cat.as_ordered()
Earthquake.set_index(['Dep Levl','CSIS Levl']).sort_index()
```
